```python
df = pd.read_csv("/content/Bestsellers with categories (1).csv")
df.head(10)
```

```python
price_1 = np.quantile(df['Price'], 0.1)
```

```python
price_1
```

```python
price_q1 = np.quantile(df['Price'],0.25)
price_q2 = np.quantile(df['Price'],0.50)
price_q3 = np.quantile(df['Price'],0.75)
```

```python
print("Price Quartiles")
print("Q1 - ",price_q1)
print("Q2 - ",price_q2)
print("Q3 - ",price_q3)
```

```python
IQR_age = price_q3 - price_q1
print("interquartile range - ",IQR_age)
```

```python
userrate_q1 = np.quantile(df['User Rating'],0.25)
userrate_q2 = np.quantile(df['User Rating'],0.50)
userrate_q3 = np.quantile(df['User Rating'],0.75)
```

```python
print("user rating quartiles")
print("Q1 - ",userrate_q1)
print("Q2 - ",userrate_q2)
print("Q3 - ",userrate_q3)
```

```python
IQR_rate = userrate_q3 - userrate_q1
print("The interquartile range -",IQR_rate)
```

```python
sns.boxplot(x="Price", data=df)
plt.title("A boxplot about the price and user rating")
```
